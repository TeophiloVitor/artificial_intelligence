```python
import numpy as np #libraries
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import math
```

```python
def z(x,y):
    return (x**2 + y**2 + 2*x*y*math.cos(math.pi*x*y)+x+y-1) #requested function
```

```python
x = np.arange(-1,1,0.05) #function range
xy = [(j,k) for j in x for k in x]
out = [z(p[0],p[1]) for p in xy]
```

```python
x_train, x_test, y_train, y_test = train_test_split(xy, out) #training and testing definition
```

```python
#mse array
arr = np.empty(5)
```

```python
#set up network with parameters
mlp = MLPRegressor(
      hidden_layer_sizes=[5],
      max_iter=1000,
      tol=0,
  )

mlp.fit(x_train,y_train) #train network

predictions = mlp.predict(x_test) #test

mse = mean_squared_error(y_test, predictions)
```

```python
y = np.array([1.3438102594204995, 0.7924179590331073, 0.1731673482314536, 0.044706593435991114]) #20 neuronios
x = np.array([1, 10, 100, 1000]) #20 neuronios
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(x,y) 

plt.title("MSE com 20 neurônios e 1000 épocas") 

plt.xlabel("epochs") 
plt.ylabel("MSE") 

plt.show()
```
